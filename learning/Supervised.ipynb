```python
import numpy as np
import sklearn

%matplotlib widget
import matplotlib.pyplot as plt
```

```python
dataset = sklearn.datasets.load_digits()

Xtrain, Xtest, ytrain, ytest = sklearn.model_selection.train_test_split(
    dataset.data, dataset.target,
    test_size=0.2)

Xtrain.shape, Xtest.shape
```

```python
H, W = 5, 5
fig,axs = plt.subplots(H, W)
k = 0
for i in range(H):
    for j in range(W):
        axs[i,j].imshow(dataset['images'][k])
        axs[i,j].axis('off')
        k += 1
```

```python
def show_evaluation(model, X, y):
    out = model.predict(X)
    (idxs, counts) = np.unique( np.vstack([y,out]), return_counts=True, axis=1 )
    confusion = np.zeros((10, 10))
    confusion[idxs[0], idxs[1]] = counts
    print(confusion)
    acc = 100 * np.sum(y==out) / len(y)
    print(f"Accuracy: {acc}%")
```

```python
model = sklearn.neighbors.KNeighborsClassifier(n_neighbors=1)
model.fit(Xtrain, ytrain)

print("----- PERFORMANCE ON THE TRAIN SET -----")
show_evaluation(model, Xtrain, ytrain)

print("\n----- PERFORMANCE ON THE TEST SET -----")
show_evaluation(model, Xtest, ytest)
```
