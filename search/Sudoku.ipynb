**Problem statement**: given a partially filled 9x9 board, find all remaining values (1-9) such that:
- no repetitions at each row
- no repetitions at each column
- no repetitions at each 3x3 subgrids

```python
import search
```

Some convenience functions to make the state space definition more readable.

```python
# Loads a board from a string (line separated rows; space separated columns)
def load_board(board):
    return [ [ int(a) for a in ln.split() ]
             for ln in board.split("\n") if len(ln.split())>0 ]

# Prints a board
def print_board(board):
    for row in board:
        print(' '.join(map(str, row)))

# Finds the coordinates of the first free slot on the board
def find_free_slot(board):
    for (i,row) in enumerate(board):
        for (j,value) in enumerate(row):
            if value==0:
                return (i, j)
    return None

# Checks the validity of placing n on the coordinates (i,j)
def valid(n, i, j, board):
    if n in board[i]:
        return False
    col = [row[j] for row in board]
    if n in col:
        return False
    oi = 3*(i//3)
    oj = 3*(j//3)
    l = [ board[oi+di][oj+dj] for di in range(3) for dj in range(3) ]
    if n in l:
        return False
    return True
```

### State space definition

The states are boards, represented as a list of lists of numbers, and an action is a tuple (i,j,n) corresponding to placing n on cell (i,j)

The following class defines:
- initialization by loading a board from a string
- the actions possible given a state -- valid numbers on the first empty cell
- the transitions -- placing a number on a cell
- the goal test -- whether there are no empty cells

```python
class Sudoku(search.Problem):

    def __init__(self, board_text):
        self.initial = load_board(board_text)

    def actions(self, state):
        (i, j) = find_free_slot(state)
        return [(i,j,n) for n in range(1, 10) if valid(n, i, j, state)]

    def result(self, state, action):
        (i,j,n) = action
        new = list(state)
        new[i] = list(state[i])
        new[i][j] = n
        return new

    def goal_test(self, state):
        return all( (0 not in row for row in state) )
```
