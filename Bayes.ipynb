```python
import ipywidgets as widgets
from IPython.display import display
```

Bayes rule illustration:

$$P(\text{covid}\,|\,\text{fever}) = \frac{P(\text{fever}\,|\,\text{covid}) P(\text{covid})}{P(\text{fever}\,|\,\text{covid}) P(\text{covid}) + P(\text{fever}\,|\,\neg\text{covid}) P(\neg\text{covid})}$$

```python
pc_slider = widgets.FloatSlider(value=0.1, min=0, max=1.0, step=0.01, description='P(covid)')
pfc_slider = widgets.FloatSlider(value=0.2, min=0, max=1.0, step=0.01, description='P(fever|covid)')
pfnc_slider = widgets.FloatSlider(value=0.05, min=0, max=1.0, step=0.01, description='P(fever|-covid)')

pcf_bar = widgets.FloatProgress(value=0, min=0, max=1.0, description='P(covid|fever)')

display(pc_slider, pfc_slider, pfnc_slider, pcf_bar)
```

```python
def recompute(_):
    pc = pc_slider.value
    pfc = pfc_slider.value
    pfnc = pfnc_slider.value
    # Bayes rule
    pcf = pfc*pc / (pfc*pc + pfnc*(1-pc))
    pcf_bar.value = pcf

recompute(_)
pc_slider.observe(recompute)
pfc_slider.observe(recompute)
pfnc_slider.observe(recompute)
```
