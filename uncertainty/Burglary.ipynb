### Probabilistic inference examples

P(Burglary | JohnCalls=True, MaryCalls=False)

```python
query = elimination_ask('Burglary', {'JohnCalls':True, 'MaryCalls':False}, burglary)
query[True]
```

P(Burglary | JohnCalls=True, MaryCalls=True)

```python
query = elimination_ask('Burglary', {'JohnCalls':True, 'MaryCalls':True}, burglary)
query[True]
```

P(Burglary | JohnCalls=False, MaryCalls=False)

```python
query = elimination_ask('Burglary', {'JohnCalls':False, 'MaryCalls':False}, burglary)
query[True]
```

P(Burglary | Alarm=True)

```python
query = elimination_ask('Burglary', {'Alarm':True}, burglary)
query[True]
```

P(JohnCalls | Burglary=True) and P(MaryCalls | Burglary=True)

```python
query1 = elimination_ask('JohnCalls', {'Burglary':True}, burglary)
query2 = elimination_ask('MaryCalls', {'Burglary':True}, burglary)
query1[True], query2[True]
```

P(JohnCalls | Earthquake=True) and P(MaryCalls | Earthquake=True)

```python
query1 = elimination_ask('JohnCalls', {'Earthquake':True}, burglary)
query2 = elimination_ask('MaryCalls', {'Earthquake':True}, burglary)
query1[True], query2[True]
```
