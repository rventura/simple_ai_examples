```python
import ipywidgets as widgets
from IPython.display import display
```

Bayes rule applied to the COVID and fever example:
- P(covid|fever) -- probability of having COVID given that has fever -- the **posterior**
- P(fever|covid) -- probability of having fever given that one has COVID
- P(fever|¬covid) -- probability of having fever given that one is healthy
- P(covid) -- probability of having COVID (without any other information) -- the **prior**

$$P(\text{covid}\,|\,\text{fever}) = \frac{P(\text{fever}\,|\,\text{covid}) P(\text{covid})}{P(\text{fever}\,|\,\text{covid}) P(\text{covid}) + P(\text{fever}\,|\,\neg\text{covid}) P(\neg\text{covid})}$$

Notes:
- $P(\neg\text{covid})=1-P(\text{covid})$
- $1/\left[P(\text{fever}\,|\,\text{covid}) P(\text{covid}) + P(\text{fever}\,|\,\neg\text{covid}) P(\neg\text{covid})\right]$ is a normalization constant

```python
# This cell setups the interface

pc_slider = widgets.FloatSlider(value=0.1, min=0, max=1.0, step=0.01, description='P(covid)')
pfc_slider = widgets.FloatSlider(value=0.2, min=0, max=1.0, step=0.01, description='P(fever|covid)')
pfnc_slider = widgets.FloatSlider(value=0.05, min=0, max=1.0, step=0.01, description='P(fever|-covid)')

pcf_bar = widgets.FloatProgress(value=0, min=0, max=1.0, description='P(covid|fever)')

display(pc_slider, pfc_slider, pfnc_slider, pcf_bar)
```

```python
# This this cell to activate the demo

def recompute(_):
    pc = pc_slider.value
    pfc = pfc_slider.value
    pfnc = pfnc_slider.value
    # Bayes rule
    pcf = pfc*pc / (pfc*pc + pfnc*(1-pc))
    pcf_bar.value = pcf

recompute(_)
pc_slider.observe(recompute)
pfc_slider.observe(recompute)
pfnc_slider.observe(recompute)
```
